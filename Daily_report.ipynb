```python
import pandas as pd
from datetime import datetime
```

```python
metrics = pd.read_csv("Daily_raw.csv")
```

```python
metrics.info()
```

```python
metrics[
    ["Break (t)", "Lunch (t)", "Submitting Lead (t)", "Bathroom (t)", "Ready:Wrap Time"]
    ]  = "00:" + metrics[
        ["Break (t)", "Lunch (t)", "Submitting Lead (t)", "Bathroom (t)", "Ready:Wrap Time"]]

metrics
```

```python
metrics[["Ready (t)", "Break (t)", "Lunch (t)", "Ready:Talk Time"]] = metrics[
    ["Ready (t)", "Break (t)", "Lunch (t)", "Ready:Talk Time"]].apply(
        lambda x: x.replace({"hours": ":"}, regex=True)
    )

metrics[["Ready (t)", "Break (t)", "Lunch (t)", "Ready:Talk Time"]] = metrics[
    ["Ready (t)", "Break (t)", "Lunch (t)", "Ready:Talk Time"]].apply(
        lambda x: x.replace({"min.": ":00"}, regex=True)
    )


metrics
```

```python
metrics[["Break (t)", "Lunch (t)", "Submitting Lead (t)", "Bathroom (t)", "Ready:Wrap Time"]
        ] = metrics[
    ["Break (t)", "Lunch (t)", "Submitting Lead (t)", "Bathroom (t)", "Ready:Wrap Time"]
        ].apply(lambda x: x.replace({"min.": ":"}, regex=True))
```

```python
metrics
```

```python
metrics[["Break (t)", "Lunch (t)", "Submitting Lead (t)", "Bathroom (t)", "Ready:Wrap Time"]
        ] = metrics[
    ["Break (t)", "Lunch (t)", "Submitting Lead (t)", "Bathroom (t)", "Ready:Wrap Time"]
        ].apply(lambda x: x.replace({"s.": ""}, regex=True))
```

```python
metrics
```

```python
metrics[["Ready (t)", "Break (t)", "Lunch (t)"]] = metrics[
    ["Ready (t)", "Break (t)", "Lunch (t)"]].apply(
        lambda x: x.replace({"00:1 : 1 :00": "01:01:00"}, regex=True))

metrics[["Ready (t)", "Break (t)", "Lunch (t)"]] = metrics[
    ["Ready (t)", "Break (t)", "Lunch (t)"]].apply(
        lambda x: x.replace({"00:1 : 2 :00": "01:00:00"}, regex=True))

metrics[["Ready (t)", "Break (t)", "Lunch (t)"]] = metrics[
    ["Ready (t)", "Break (t)", "Lunch (t)"]].apply(
        lambda x: x.replace({"00:60 :00 0": "01:00:00"}, regex=True))
```

```python
metrics
```

```python
metrics["Ready (t)"] = metrics["Ready (t)"].str.replace(" ", "")
metrics["Break (t)"] = metrics["Break (t)"].str.replace(" ", "")
metrics["Lunch (t)"] = metrics["Lunch (t)"].str.replace(" ", "")
metrics["Submitting Lead (t)"] = metrics["Submitting Lead (t)"].str.replace(" ", "")
metrics["Bathroom (t)"] = metrics["Bathroom (t)"].str.replace(" ", "")
metrics["DEAD TIME (t)"] = metrics["DEAD TIME (t)"].str.replace(" ", "")
metrics["Ready:Talk Time"] = metrics["Ready:Talk Time"].str.replace(" ", "")
metrics["Ready:Wrap Time"] = metrics["Ready:Wrap Time"].str.replace(" ", "")


metrics
```

```python
metrics[
    ["Ready (t)", "Break (t)", "Lunch (t)", "Submitting Lead (t)", 
    "Bathroom (t)", "DEAD TIME (t)", "Ready:Talk Time", "Ready:Wrap Time"]
] = metrics[
    ["Ready (t)", "Break (t)", "Lunch (t)", "Submitting Lead (t)", 
    "Bathroom (t)", "DEAD TIME (t)", "Ready:Talk Time", "Ready:Wrap Time"]
].apply(
        lambda x: x.replace({"00:-": "00:00:00", "-": "00:00:00"}, regex=True))
```

```python
metrics["DEAD TIME (t)"] = "00:" + metrics["DEAD TIME (t)"].astype("object")
```

```python
metrics["DEAD TIME (t)"] = metrics["DEAD TIME (t)"].str.replace("min.", ":")

metrics
```

```python
metrics["DEAD TIME (t)"] = metrics["DEAD TIME (t)"].str.replace("s.", "")
metrics["DEAD TIME (t)"] = metrics["DEAD TIME (t)"].str.replace("00:00:00:00", "00:00:00")
metrics
```

```python
for row in metrics["Submitting Lead (t)"].items():
    if len(row) == [3,4,5]:
        print("A huevo")
```

```python
metrics["Submitting Lead (t)"] = metrics["Submitting Lead (t)"].apply(
        lambda x: x if len(x) > 5 else '00:' + x)

metrics["Bathroom (t)"] = metrics["Bathroom (t)"].apply(
        lambda x: x if len(x) > 5 else '00:' + x)

metrics["DEAD TIME (t)"] = metrics["DEAD TIME (t)"].apply(
        lambda x: x if len(x) > 5 else '00:' + x)
```

```python
metrics
```

```python
metrics[["Submitting Lead (t)", "Bathroom (t)", "DEAD TIME (t)"]] = metrics[
    ["Submitting Lead (t)", "Bathroom (t)", "DEAD TIME (t)"]].astype("datetime64[s]")
```

```python
metrics["DEAD TIME (t)"] = pd.to_datetime(metrics["DEAD TIME (t)"], format='%H:%M:%S')

metrics.index = metrics.index +1
```

```python
metrics
```

```python
metrics["PENDING"] = pd.Series(dtype='int')
```

```python
metrics["DAYS WORKED"] = pd.Series(dtype='int')

metrics
```

```python
#AGENT / DAYS WORKED /TIME READY /PENDING APP /BREAK / LUNCH /SUBMIT LEAD /BATHROOM /DEAD TIME / TALK TIME /LOGGED CALLS / WRAP TIME

metrics = metrics[["Name?", "DAYS WORKED", "Ready (t)", "PENDING", "Break (t)", "Lunch (t)",
                    "Submitting Lead (t)", "Bathroom (t)", "DEAD TIME (t)", "Ready:Talk Time",
                    "Logged calls", "Ready:Wrap Time"]]

metrics

```

```python
metrics["Ready:Talk Time"] = metrics["Ready:Talk Time"].apply(
        lambda x: x if len(x) < 8 else '00:' + x)

metrics["Ready:Talk Time"] = metrics["Ready:Talk Time"].str.replace("s.", "")

```

```python
metrics
```

```python
metrics["Ready (t)"] = metrics["Ready (t)"].apply(
        lambda x: x if len(x) < 8 else '00:' + x)

metrics["Ready (t)"] = metrics["Ready (t)"].str.replace("s.", "")
```
